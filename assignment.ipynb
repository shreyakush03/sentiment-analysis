```python
##Check date ranges
print("Historical Data Date Range:")
print(f"  Start: {historical_data_df['Date'].min()}")
print(f"  End: {historical_data_df['Date'].max()}")
print(f"\nFear & Greed Date Range:")
print(f"  Start: {fear_greed_df['date'].min()}")
print(f"  End: {fear_greed_df['date'].max()}")
print(f"\nCommon date range:")
common_start = max(historical_data_df['Date'].min(), fear_greed_df['date'].min())
common_end = min(historical_data_df['Date'].max(), fear_greed_df['date'].max())
print(f"  Start: {common_start}")
print(f"  End: {common_end}")
```

```python
##Aggregate trading metrics by date
daily_trades = historical_data_df.groupby('Date').agg({
    'size usd': ['sum', 'mean', 'count'],
    'closed pnl': ['sum', 'mean'],
    'side': lambda x: (x == 'BUY').sum()
}).reset_index()

daily_trades.columns = ['Date', 'total_volume_usd', 'avg_trade_size', 'num_trades', 
                        'total_pnl', 'avg_pnl', 'buy_count']
daily_trades['sell_count'] = daily_trades['num_trades'] - daily_trades['buy_count']
daily_trades['buy_ratio'] = daily_trades['buy_count'] / daily_trades['num_trades']

print("Daily Trading Aggregates:")
print(daily_trades.head(10))
```

```python
##Merge datasets on date
merged_df = pd.merge(
    daily_trades, 
    fear_greed_df[['date', 'value', 'classification']], 
    left_on='Date', 
    right_on='date', 
    how='inner'
)
merged_df = merged_df.drop('date', axis=1)
merged_df.columns = ['Date', 'total_volume_usd', 'avg_trade_size', 'num_trades', 
                     'total_pnl', 'avg_pnl', 'buy_count', 'sell_count', 'buy_ratio',
                     'fear_greed_value', 'sentiment_class']

print(f"Merged dataset shape: {merged_df.shape}")
print(f"\nMerged Data Sample:")
print(merged_df.head(10))
```

```python
##Calculate correlations
print("CORRELATION ANALYSIS: Trading Metrics vs Fear & Greed Index")
print("="*60)

correlations = {
    'Total Volume USD': merged_df['total_volume_usd'].corr(merged_df['fear_greed_value']),
    'Avg Trade Size': merged_df['avg_trade_size'].corr(merged_df['fear_greed_value']),
    'Number of Trades': merged_df['num_trades'].corr(merged_df['fear_greed_value']),
    'Total PnL': merged_df['total_pnl'].corr(merged_df['fear_greed_value']),
    'Avg PnL': merged_df['avg_pnl'].corr(merged_df['fear_greed_value']),
    'Buy Ratio': merged_df['buy_ratio'].corr(merged_df['fear_greed_value']),
}

for metric, corr in correlations.items():
    print(f"{metric:.<30} {corr:>7.4f}")

print("\nInterpretation:")
print("  >  0.3:  Positive correlation (high sentiment → higher metric)")
print("  < -0.3:  Negative correlation (high sentiment → lower metric)")
print("  [-0.3, 0.3]: Weak/no correlation")
```

```python
##Visualizations: Relationship between Trading & Fear/Greed
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Relationship: Trading Metrics vs Fear & Greed Index', fontsize=16)

axes[0, 0].scatter(merged_df['fear_greed_value'], merged_df['total_volume_usd'], alpha=0.5)
axes[0, 0].set_xlabel('Fear & Greed Index')
axes[0, 0].set_ylabel('Total Volume USD')
axes[0, 0].set_title(f"Volume (r={correlations['Total Volume USD']:.3f})")

axes[0, 1].scatter(merged_df['fear_greed_value'], merged_df['num_trades'], alpha=0.5)
axes[0, 1].set_xlabel('Fear & Greed Index')
axes[0, 1].set_ylabel('Number of Trades')
axes[0, 1].set_title(f"Trade Count (r={correlations['Number of Trades']:.3f})")

axes[0, 2].scatter(merged_df['fear_greed_value'], merged_df['buy_ratio'], alpha=0.5)
axes[0, 2].set_xlabel('Fear & Greed Index')
axes[0, 2].set_ylabel('Buy Ratio')
axes[0, 2].set_title(f"Buy Ratio (r={correlations['Buy Ratio']:.3f})")

axes[1, 0].scatter(merged_df['fear_greed_value'], merged_df['avg_trade_size'], alpha=0.5)
axes[1, 0].set_xlabel('Fear & Greed Index')
axes[1, 0].set_ylabel('Avg Trade Size USD')
axes[1, 0].set_title(f"Avg Trade Size (r={correlations['Avg Trade Size']:.3f})")

axes[1, 1].scatter(merged_df['fear_greed_value'], merged_df['total_pnl'], alpha=0.5)
axes[1, 1].set_xlabel('Fear & Greed Index')
axes[1, 1].set_ylabel('Total PnL')
axes[1, 1].set_title(f"Total PnL (r={correlations['Total PnL']:.3f})")

axes[1, 2].scatter(merged_df['fear_greed_value'], merged_df['avg_pnl'], alpha=0.5)
axes[1, 2].set_xlabel('Fear & Greed Index')
axes[1, 2].set_ylabel('Avg PnL per Trade')
axes[1, 2].set_title(f"Avg PnL (r={correlations['Avg PnL']:.3f})")

plt.tight_layout()
plt.show()
```

```python
##Analysis by Sentiment Classification
print("\nMETRICS BY SENTIMENT CLASSIFICATION")
print("="*60)

sentiment_analysis = merged_df.groupby('sentiment_class').agg({
    'total_volume_usd': ['mean', 'std'],
    'num_trades': ['mean', 'std'],
    'buy_ratio': ['mean', 'std'],
    'total_pnl': ['mean', 'std'],
    'fear_greed_value': ['mean']
}).round(2)

print(sentiment_analysis)

print("\n\nSUMMARY INSIGHTS:")
print("-" * 60)
fear_avg_volume = merged_df[merged_df['sentiment_class'].isin(['Fear', 'Extreme Fear'])]['total_volume_usd'].mean()
greed_avg_volume = merged_df[merged_df['sentiment_class'].isin(['Greed', 'Extreme Greed'])]['total_volume_usd'].mean()
print(f"Average Daily Volume during FEAR periods:  ${fear_avg_volume:,.0f}")
print(f"Average Daily Volume during GREED periods: ${greed_avg_volume:,.0f}")
print(f"Difference: {((greed_avg_volume - fear_avg_volume) / fear_avg_volume * 100):+.1f}%")

fear_buy_ratio = merged_df[merged_df['sentiment_class'].isin(['Fear', 'Extreme Fear'])]['buy_ratio'].mean()
greed_buy_ratio = merged_df[merged_df['sentiment_class'].isin(['Greed', 'Extreme Greed'])]['buy_ratio'].mean()
print(f"\nBuy Ratio during FEAR periods:  {fear_buy_ratio:.1%}")
print(f"Buy Ratio during GREED periods: {greed_buy_ratio:.1%}")
```
